# Merged, data cleaning and add calculation fields etc.

```python
exp_post_big = pd.read_excel('/content/drive/MyDrive/Expedia_AspireData_Cleaning_Nov23/Expedia_Big_Dataset_Nov_1115.xlsx')
```

```python
# filter only needed data: remove YTD
exp_post_big = exp_post_big.filter(regex='^(?!YTD)')
```

```python
# member id 42601807 has missing influ's info, aligned with id 42601558
row_copy_value = exp_post_big.loc[exp_post_big['member_id'] == 42601558, exp_post_big.columns[18:35]].values[0]

# insert values where member_id is 42601807 with values from 42601558
exp_post_big.loc[exp_post_big['member_id'] == 42601807, exp_post_big.columns[18:35]] = row_copy_value
```

```python
# Dionna has missing brand value
exp_post_big.loc[(exp_post_big['member_name'] == 'Dionna Hickox | Moments In Midlife') & (exp_post_big['brand'].isnull()), 'brand'] = 'Expedia'
```

```python
exp_post_big['comment'] = pd.to_numeric(exp_post_big['comment'])
exp_post_big['tot_likes'] = pd.to_numeric(exp_post_big['tot_likes'])
exp_post_big['tot_shares'] = pd.to_numeric(exp_post_big['tot_shares'])
```

```python
# fill 0 in null "comment" column value
exp_post_big['comment'] = exp_post_big['comment'].fillna(0)
exp_post_big['tot_likes'] = exp_post_big['tot_likes'].fillna(0)
exp_post_big['tot_shares'] = exp_post_big['tot_shares'].fillna(0)
```

```python
# remove rows that member name is null
exp_post_big = exp_post_big.dropna(subset=['member_name'])
```

```python
# calculate Engagement (Like + Share + Comments)
exp_post_big['Engagement'] = exp_post_big['comment'] + exp_post_big['tot_likes'] + exp_post_big['tot_shares']
```

```python
# calculate ER (Eng/current_followers)
exp_post_big['ER'] = np.select(
    [
        exp_post_big['network'] == 'instagram',
        exp_post_big['network'] == 'tiktok',
        exp_post_big['network'] == 'youtube'
    ],
    [
        exp_post_big['Engagement'] / exp_post_big['ig_followers'],
        exp_post_big['Engagement'] / exp_post_big['tt_followers'],
        exp_post_big['Engagement'] / exp_post_big['yt_subscribers']
    ],
    default=pd.NA  # fill in nan to null
)
```

```python
# calculate VVR (tot_views/imp)
# calculate TT, YT VVR (tt and yt has no imp so use followers to count)
exp_post_big['VVR'] = np.select(
    [
        exp_post_big['network'] == 'instagram',
        exp_post_big['network'] == 'tiktok',
        exp_post_big['network'] == 'youtube'
    ],
    [
        exp_post_big['tot_views'] / exp_post_big['tot_impressions'],
        exp_post_big['tot_views'] / exp_post_big['tt_followers'],
        exp_post_big['tot_views'] / exp_post_big['yt_subscribers']
    ],
    default=pd.NA  # fill in nan to null
)
```

```python
#exp_post_big.iloc[35:40, -2]
```

```python
# change to %
#exp_post_big['ER'] = exp_post_big['ER'].apply(lambda x: '{:.0%}'.format(x))
```

```python
# calculate SRR (imp/current_followers)
exp_post_big.loc[exp_post_big['post_type'] == 'instagram_story', 'SRR'] = exp_post_big['tot_impressions'] / exp_post_big['current_followers']
# fill NaN for rows where post_type is not instagram_story
exp_post_big['SRR'].fillna(value=pd.NA, inplace=True)
```

```python
def capitalize_column_name(column_name):
    if '_' in column_name:
        # Capitalize first letter and first letter after _
        words = column_name.split('_')
        capitalized_words = [word.capitalize() for word in words]
        return '_'.join(capitalized_words)
    else:
        # Capitalize both words if there are spaces
        return ' '.join(word.capitalize() for word in column_name.split())

# Apply the function to each column name
exp_post_big.columns = [capitalize_column_name(column) for column in exp_post_big.columns]
```
